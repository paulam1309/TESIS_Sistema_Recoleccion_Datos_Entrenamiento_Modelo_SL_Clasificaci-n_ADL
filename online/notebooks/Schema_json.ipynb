```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd, json, os

ruta_ventana_50_sin_balanceo = "/content/drive/MyDrive/Colab Notebooks/TESIS/FASE OFFLINE/DATASET_ETAPAS_POSTERIORES/ventanas_mu√±eca_50.csv"
df = pd.read_csv(ruta_ventana_50_sin_balanceo)

FEATURES = [c for c in df.columns if c not in ['etiqueta','start_index','end_index','n_muestras']]
schema = {"features": FEATURES}

# Carpeta destino en Drive
carpeta_guardar_schema = "/content/drive/MyDrive/Colab Notebooks/TESIS/FASE ONLINE/ARCHIVOS ONLINE"
os.makedirs(carpeta_guardar_schema, exist_ok=True)

# Archivo final
nombre_archivo = os.path.join(carpeta_guardar_schema, "schema.json")
with open(nombre_archivo, "w", encoding="utf-8") as f:
    json.dump(schema, f, ensure_ascii=False, indent=2)

print("Guardado en:", nombre_archivo)
print("Features:", len(FEATURES))
```
